```python
import gradio as gr
import requests

BACKEND_URL = "http://localhost:8000/process"  # Change to your backend URL

def process_input(user_text):
    if not user_text.strip():
        return "Please enter some text!", "", ""

    try:
        response = requests.post(BACKEND_URL, json={"user_text": user_text})
        result = response.json()

        granite_resp = result.get("granite", "No Granite response.")
        nlu_resp = result.get("nlu", "No NLU data.")
        chain_resp = result.get("chain", "No LangChain summary.")

        # Convert dict NLU output to pretty string if needed
        if isinstance(nlu_resp, dict):
            import json
            nlu_resp = json.dumps(nlu_resp, indent=2)

        return granite_resp, nlu_resp, chain_resp
    except Exception as e:
        return f"Error: {e}", "", ""

def process_granite(user_text):
    if not user_text.strip():
        return "Please enter some text!"
    try:
        response = requests.post(BACKEND_URL, json={"user_text": user_text})
        result = response.json()
        return result.get("granite", "No Granite response.")
    except Exception as e:
        return f"Error: {e}"

def process_nlu(user_text):
    if not user_text.strip():
        return "Please enter some text!"
    try:
        response = requests.post(BACKEND_URL, json={"user_text": user_text})
        result = response.json()
        nlu_resp = result.get("nlu", "No NLU data.")
        if isinstance(nlu_resp, dict):
            import json
            nlu_resp = json.dumps(nlu_resp, indent=2)
        return nlu_resp
    except Exception as e:
        return f"Error: {e}"

def process_chain(user_text):
    if not user_text.strip():
        return "Please enter some text!"
    try:
        response = requests.post(BACKEND_URL, json={"user_text": user_text})
        result = response.json()
        return result.get("chain", "No LangChain summary.")
    except Exception as e:
        return f"Error: {e}"

granite_interface = gr.Interface(
    fn=process_granite,
    inputs=gr.Textbox(lines=3, placeholder="Type your finance question here..."),
    outputs="text",
    title="Granite AI Response",
    description="Get response from Granite model."
)

nlu_interface = gr.Interface(
    fn=process_nlu,
    inputs=gr.Textbox(lines=3, placeholder="Type your finance question here..."),
    outputs="text",
    title="Watson NLU Insights",
    description="Get NLU analysis."
)

chain_interface = gr.Interface(
    fn=process_chain,
    inputs=gr.Textbox(lines=3, placeholder="Type your finance question here..."),
    outputs="text",
    title="LangChain Summary",
    description="Get LangChain summary."
)

demo = gr.TabbedInterface(
    interface_list=[granite_interface, nlu_interface, chain_interface],
    title="üí∞ FinMate - Personal Finance Assistant"
)

demo.launch(share=True)
```
